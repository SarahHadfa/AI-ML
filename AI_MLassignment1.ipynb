```python
graph = {
    'X': ['Y', 'Z'],
    'Y': ['A', 'B'],
    'Z': ['C'],
    'A': [],
    'B': ['D'],
    'C': [],
    'D': []
}

visited = set()

def dfs(node):
    if node not in visited:
        visited.add(node)
        print(node, end=" ")

        for neighbor in graph[node]:
            dfs(neighbor)

print("DFS Traversal:")
dfs('X')
```
